```python
import time
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Retrieve the url of the 25 best performances in triple jump of the year 1996
# http://trackfield.brinkster.net/25BestPerf.asp?EventCode=MF4&P=F
# Creation of a list of URLs (men + women), a list of gender (M or W) and a list of years
year = 1891
sex = 'M'
url = []
years = []
genre = []

while year <= 2020:
    u = 'http://trackfield.brinkster.net/More.asp?Year={}&EventCode={}F4'.format(year, sex)
    url.append(u)
    years.append(year)
    genre.append(sex)
    year = year+1

year = 1891
sex = 'W'
while year <= 2020:
    u = 'http://trackfield.brinkster.net/More.asp?Year={}&EventCode={}F4'.format(year, sex)
    url.append(u)
    years.append(year)
    genre.append(sex)
    year = year+1

test = True
tab_final=[]

# Data recovery (Loop on the three list)
for y,u,s in zip(years,url,genre):
    
    # Try / except in case of server saturation
    test = True
    print(u)
    while test :
        try:
            print('try')
            x = requests.get(u)
            print('request ok')
            test = False
        except :
            print('except')
            time.sleep(2)
            test = True
    print('start of treatment')
    
    lst=[]
    
    soup = BeautifulSoup(x.content,'html.parser')

    for p in soup.find_all('td', attrs={"bgcolor":"#F3F2EE"}):
        # we do not record lines == indoor
        if p.text !=   'Indoor Results'  :
            lst.append(p.text)          
        
    # Delete the 1st line (the title)
    lst = lst[1:]   
    
    lst_temp=[]
    
    # Resize the list into columns    
    lst = np.reshape(lst, (int(len(lst)/8), 8))
    print(len(lst))
    
    # Adding a temp list then ds final tab of data + year and sex
    for lign in range(len(lst)) :        
        an = (y,s)
        lst_temp = np.append(lst[lign,:],an)          
        tab_final.append(lst_temp)


# Creation of the data frame with the different columns
df = pd.DataFrame(data=tab_final,columns=['Range','Names','Country','Performances',
                                          'Null','Nb','City','Date','Year','Sex'])

# Converting data frame to xlsx
# writer = pd.ExcelWriter(r'data/athletes.xlsx')
# df.to_excel(writer,'Sheet1')

# Converting data frame to csv
df.to_csv(r'data/athletes.csv')
```
